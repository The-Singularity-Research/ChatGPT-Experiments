```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay

# Generate a random point cloud in 2D
num_points = 50
points = np.random.rand(num_points, 2)

# Compute the Delaunay triangulation of the point cloud
tri = Delaunay(points)

# Assign random weights to the vertices, edges, and faces
vertex_weights = np.random.rand(num_points)
edge_weights = np.random.rand(tri.simplices.shape[0], 3)
face_weights = np.random.rand(tri.simplices.shape[0])

# Plot the simplicial complex with weights
plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
plt.plot(points[:,0], points[:,1], 'o')

# Label the vertices with their weights
for i, point in enumerate(points):
    plt.annotate(f'{vertex_weights[i]:.2f}', (point[0], point[1]), 
                 xytext=(-15,15), textcoords='offset points',
                 ha='center', va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                 arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

# Label the edges with their weights
for i, triangle in enumerate(tri.simplices):
    for j, vertex in enumerate(triangle):
        x1, y1 = points[vertex]
        x2, y2 = points[triangle[(j + 1) % 3]]
        plt.annotate(f'{edge_weights[i, j]:.2f}', ((x1 + x2) / 2, (y1 + y2) / 2), 
                     xytext=(-15, 15), textcoords='offset points',
                     ha='center', va='bottom',
                     bbox=dict(boxstyle='round,pad=0.5', fc='blue', alpha=0.5),
                     arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

# Label the faces with their weights
for i, triangle in enumerate(tri.simplices):
    x1, y1 = points[triangle[0]]
    x2, y2 = points[triangle[1]]
    x3, y3 = points[triangle[2]]
    plt.annotate(f'{face_weights[i]:.2f}', ((x1 + x2 + x3) / 3, (y1 + y2 + y3) / 3), 
                 xytext=(-15, 15), textcoords='offset points',
                 ha='center', va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='red', alpha=0.5),
                 arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))
    
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [24, 18]
plt.show()
```
